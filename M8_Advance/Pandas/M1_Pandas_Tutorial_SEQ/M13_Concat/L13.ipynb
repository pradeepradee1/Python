# Topic : Concat

```python
import pandas as pd
```

```python
weather = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "Temp" : [32,45,30],
    "humidity":[80,60,70]
})
weather
```

```python
weather1 = pd.DataFrame({
    "city":["new york","chicago","oriando"],
    "Temp" : [31,42,80],
    "humidity":[86,45,34]
})
weather1
```

```python
df=pd.concat([weather,weather1])
df
```

```python
df=pd.concat([weather,weather1],ignore_index=True)
df
```

# Note : Want to know more parameters in pandas go to official site

```python
df=pd.concat([weather,weather1],ignore_index=True,keys=["india","us"])
df
```

```python
df=pd.concat([weather,weather1],keys=["india","us"])
df
```

```python
df.loc["india"]
```

```python
data = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "Temp" : [32,45,30]
})
data
```

```python
data1 = pd.DataFrame({
    "city":["new york","chicago","oriando"],
    "humidity":[86,45,34]
})
data1
```

```python
data3=pd.concat([data,data1])
data3
```

```python
data3=pd.concat([data,data1],axis=1)
data3
```

## Note : Concating Two Dataframes in column wise based on the common index value in two dataframe

```python
data = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "Temp" : [32,45,30]
})
data
```

```python
data1 = pd.DataFrame({
    "city":["delhi","chicago"],
    "humidity":[7,12]
})
data1
```

```python
data3=pd.concat([data,data1],axis=1)
data3
```

```python
data = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "Temp" : [32,45,30]
},index=[0,1,2])
data
```

```python
data1 = pd.DataFrame({
    "city":["delhi","mumbai"],
    "humidity":[7,12]
},index=[1,0])
data1
```

```python
data3=pd.concat([data,data1],axis=1)
data3
```

```python
data.set_index('Temp',inplace=True)
data
```
