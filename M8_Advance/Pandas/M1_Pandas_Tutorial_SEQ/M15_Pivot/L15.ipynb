# Pivot is used for grouping the values based on the Index's feature values And Columns's feature values in pivot table function

### Pivot allows you to transform and reshape datafarme 

```python
import pandas as pd
```

```python
data=[['5/1/2017','new york',65,56],['5/2/2017','new york',66,58],['5/3/2017','new york',68,60],['5/1/2017','mumbai',75,80],['5/2/2017','mumbai',78,83],['5/3/2017','mumbai',82,85],['5/1/2017','beijing',80,2],['5/2/2017','beijing',77,30],['5/3/2017','beijing',79,35]]
df=pd.DataFrame(data,columns=['date','city','temperature','humidity'])
df
```

```python
df.pivot(index="date",columns="city")
```

```python
df.pivot(index="date",columns="city",values="humidity")
```

```python
df.pivot(index="humidity",columns="city")
```

# pivot_table is used to summarize and aggregate data inside dataframe

```python
data=[['5/1/2017','new york',65,56],['5/1/2017','new york',61,54],['5/2/2017','new york',70,60],['5/2/2017','new york',72,62],['5/1/2017','mumbai',75,80],['5/1/2017','mumbai',78,83],['5/2/2017','mumbai',80,2],['5/2/2017','mumbai',80,26]]
df=pd.DataFrame(data,columns=['date','city','temperature','humidity'])
df
```

```python
df.pivot_table(index='city',columns="date")
```

```python
df.pivot_table(index='city',columns="date",aggfunc="sum")
```

```python
df.pivot_table(index='city',columns="date",aggfunc="count")
```

```python
df.pivot_table(index='city',columns="date",aggfunc="diff")
```

```python
df.pivot_table(index='city',columns="date",aggfunc="mean")
```

```python
df.pivot_table(index='city',columns="date",margins=True)
```

```python
data=[['5/1/2017','new york',65,56],['5/2/2017','new york',61,54],['5/3/2017','new york',70,60],['12/1/2017','new york',30,50],['12/2/2017','new york',28,52],['12/3/2017','new york',28,52]]
df=pd.DataFrame(data,columns=['date','city','temperature','humidity'])
df
```

```python
df['date']=pd.to_datetime(df["date"])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns="city")
```
