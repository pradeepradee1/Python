## Melt is used to transform Or reshape data

```python
import pandas as pd
data=[['Monday',32,75,41],['Tuesday',30,77,43],['Wednesday',28,75,45],['Thursday',22,82,38],['Friday',30,83,30],['Sat',20,81,43],['Sun',25,77,47]]
df=pd.DataFrame(data,columns=['day','Chicago','Chennai','berlin'])
df
```

```python
df1=pd.melt(df,id_vars=["day"])
df1
```

```python
df1[df1["variable"]=="Chennai"]
```

```python
df1
```

```python
df1=pd.melt(df,id_vars=["day"],var_name="city",value_name="Temp")
df1
```
