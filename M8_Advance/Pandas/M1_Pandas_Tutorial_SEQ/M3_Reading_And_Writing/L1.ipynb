```python
def convert_people_cell(cell):
    if cell == "n.a." or cell == "bill gates":
        return 'Pradeep'
    return cell

def convert_eps_cell(cell):
    if cell =="not available":
        return None
    return cell
```

```python
import pandas as pd
df=pd.read_csv("data.csv",header=1,converters={"people":convert_people_cell,"eps":convert_eps_cell})
df
```
