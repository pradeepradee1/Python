## 1) Indexing based on the indexing
## 2) loc()
## 3) iloc()

```python
import pandas as pd
df = {
    "day": ["1/1/2017","1/2/2017","1/3/2017","1/4/2017","1/5/2017","1/6/2017"],
    "temperature":[32,35,28,24,32,31],
    "windspeed":[6,7,2,7,4,2],
    "event":["Rain","Sunny","Snow","Snow","Rain","Sunny"]
}
df = pd.DataFrame(df)
df
```

# 1 . Indexing based on index

```python
df[2:4]
```

```python
df[0:1]
```

# 2. Ioc()
## loc = Locations

#### 2.1 ) Row

```python
df.loc[:,:]
```

```python
df.loc[1:3,:]
```

```python
df.loc[[1,3,5],:]
```

### 2.2 )  Columns

```python
df.loc[:,:'temperature']
```

```python
df.loc[:,'temperature':'windspeed']
```

```python
df.loc[:,['day','event']]
```

# 3 . Iloc()

### 3.1 . Rows

```python
df.iloc[:,:]
```

```python
df.iloc[1:3,:]
```

```python
df.iloc[[1,3,5],:]
```

### 3.2 . Columns

```python
# df.iloc[:,:'temperature'] --->ERROR
df.iloc[:,:2]
```

```python
df.iloc[:,2:3]
```

```python
df.iloc[:,[2,3]]
```
