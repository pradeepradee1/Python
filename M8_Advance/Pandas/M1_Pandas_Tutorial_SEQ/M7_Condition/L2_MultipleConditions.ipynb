```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 70000, 80000]}
df = pd.DataFrame(data)

# Create a new column 'Seniority' based on the Age
df.loc[df['Age'] < 30, 'Seniority'] = 'Junior'
df.loc[(df['Age'] >= 30) & (df['Age'] < 40), 'Seniority'] = 'Mid-Level'
df.loc[df['Age'] >= 40, 'Seniority'] = 'Senior'
df
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 70000, 80000]}
df = pd.DataFrame(data)

# Create a new column 'Seniority' based on the Age
df['Seniority'] = np.where(df['Age'] < 30, 'Junior', np.where(df['Age'] < 40, 'Mid-Level', 'Senior'))
df
```
